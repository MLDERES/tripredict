```python
from bs4 import BeautifulSoup
import re

results_urls = []
competitions_url = 'https://www.endurance-data.com/en/competitions/'

for pg in range(1,20):
    r = requests.get(f'{competitions_url}{pg}', headers=header)
    soup = BeautifulSoup(r.text)
    for race in soup.find_all('a', class_='btn'):
        href = race['href']
        if (re.search('703.*/all', href)):
            results_urls.append(href[1:])

ds = pd.Series(results_urls)
ds.to_csv('../data/results_urls.txt')
```

```python

def get_race_name_from_url(url):
    m = re.search(r'(ironman.*)\/all',url)
    return(m[1])

def get_race_id_from_url(url):
    m = re.search(r'(\d{1,3})-ironman',url)
    return(m[1])

get_race_name_from_url('en/results/6-ironman-703-zell-am-see/all/')
get_race_id_from_url('en/results/6-ironman-703-zell-am-see/all/')
```

```python
base_race_url = 'http://www.endurance-data.com/'
ds_complete_results = []
#for race_url in ds:
for race_url in ds:
    print(f'Currently checking {race_url}')
    more_pages = True
    ds_results = []
    results_page = 0
    while (more_pages):
        results_page += 1
        try:
            req_url = f'{base_race_url}{race_url}{results_page}'
            print(req_url)
            r = requests.get(req_url)
            ds_race_results = pd.read_html(r.text,header=0)[0]
            ds_race_results.rename(columns={'Ovr':'OverallRank',
                                       'Gen':'GenderRank',
                                       'Div':'DivRank',
                                       '#':'Bib',
                                       'AG':'AgeGroup',
                                       'Unnamed: 6':'Country',
                                       'Unnamed: 7':"SwimTime",
                                       'Unnamed: 8':'BikeTime',
                                       'Unnamed: 9':'RunTime',
                                       'Unnamed: 10':'Overall'},
                              inplace=True)
            ds_race_results['Race'] = get_race_name_from_url(race_url)
            ds_results.append(ds_race_results)
        except ValueError:
            more_pages=False
        
        print(f'{len(ds_results)}')
        ds_onerace = ds_results.pop()
        ds_onerace = ds_onerace.append(ds_onerace,ignore_index=True)
        ds_onerace.to_csv(f'../data/{get_race_id_from_url(race_url)}.csv')
            
# ds_final = ds_complete_results.pop()
# ds_final = ds_final.append(ds_complete_results,ignore_index=True)
```
